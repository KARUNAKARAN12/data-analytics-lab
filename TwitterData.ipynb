# Tokenization and stemming

```
import pandas as pd
df=pd.read_csv("tweets.csv")
from nltk.stem import PorterStemmer
from nltk.tokenize import sent_tokenize, word_tokenize
import nltk
```

```
nltk.download('punkt')
```

```
ps = PorterStemmer()
allTweets =df['tweet']
stemmedWords = []
for tweet in allTweets:
  wordsInTweet = word_tokenize(tweet)
  for word in wordsInTweet:
    stemmedWords.append(ps.stem(word))
print (stemmedWords)
```

```
import re
word_alpha = []
for w in stemmedWords:
  match = re.search(r'[A-Za-z]+',w)
  if match:
    word_alpha.append(w)
word_alpha
```

```
from nltk import ngrams


threegrams = ngrams(word_alpha, 3)

for grams in threegrams:
  print(grams)
```

```
#document term matrix
from sklearn.feature_extraction.text import CountVectorizer
vec = CountVectorizer()
X = vec.fit_transform(allTweets)
df_term = pd.DataFrame(X.toarray(), columns=vec.get_feature_names())
print(df_term)
```
